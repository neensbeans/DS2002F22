```python
# import relevant packages
import os
import pandas as pd
import pymongo
import json
```

```python
#note: zip file names were changed as follows: Best Movie by Year Netflix -> bestmoviebyyear; Best Movies Netflix -> bestmovies
#Best Show by Year Netflix -> bestshowbyyear; Best Shows Netflix -> bestshows

#read in all the csv files
os.chdir(r'/Users/ninij/Desktop/NetflixData')
bestMovieByYear = pd.read_csv(r'bestmoviebyyear.csv')
bestMovies = pd.read_csv(r'bestmovies.csv')
bestShowByYear =  pd.read_csv(r'bestshowbyyear.csv')
bestShows =  pd.read_csv(r'bestshows.csv')
rawCredits =  pd.read_csv(r'raw_credits.csv')
rawTitles =  pd.read_csv(r'raw_titles.csv')
```

```python
bestMovieByYear.drop('index', inplace=True, axis=1)
bestMovieByYear
```

```python
bestMovies.drop('index', inplace=True, axis=1)
bestMovies
```

```python
bestShowByYear.drop('index', inplace=True, axis=1)
bestShowByYear
```

```python
bestShows.drop('index', inplace=True, axis=1)
bestShows
```

```python
rawCredits.drop('index', inplace=True, axis=1)
rawCredits
```

```python
rawTitles.drop('index', inplace=True, axis=1)
rawTitles
ageCert = rawTitles[['title', 'age_certification']].copy()
ageCert.rename(columns={"title": "TITLE", "age_certification": "AGE_CERTIFICATION"}, inplace=True)
ageCert
```

```python
#create final mongo movie collection
mergedMovieData = bestMovies.merge(ageCert, how='inner', on='TITLE')
mergedMovieData.drop('NUMBER_OF_VOTES', inplace=True, axis=1)
```

```python
#create final mongo movie collection
mergedShowData = bestShows.merge(ageCert, how='inner', on='TITLE')
mergedShowData.drop('NUMBER_OF_VOTES', inplace=True, axis=1)
```

```python
# final pandas dataframes to be uploaded to mongoDB
mergedMovieData
```

```python
mergedShowData
```

```python
# Moving data to mongoDB
host_name = "localhost"
port = "27017"

conn_str = {
    "local" : f"mongodb://{host_name}:{port}/",
}
client = pymongo.MongoClient(conn_str["local"])

print(f"Local Connection String: {conn_str['local']}")
db_name = "Netflix"
db = client[db_name]
```

```python
# make collection for movie data:
MovieData = mergedMovieData.to_dict(orient = "records")
db.Movies.insert_many(MovieData)
```

```python
# make collection for show data: 
ShowData = mergedShowData.to_dict(orient = "records")
db.Shows.insert_many(ShowData)
```
