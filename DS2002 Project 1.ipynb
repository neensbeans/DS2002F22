```python
# These lines of codes read in the data in the thyroid disease csv file I will be using for my project.
data_file = os.path.join(r'/Users/ninij/Desktop/thyroidDF.csv')  # based on where the user has uploaded the thyroidDF.csv file, the path name will need to be modified.
thyroidData = pd.read_csv(data_file, header=0, index_col=30) # sets up the csv index based on the patient IDs
thyroidData.head()
```

```python
# For this project, I will be removing all the patient queries from the table. This includes query_on_thyroxine, query_hypothyroid, query_hyperthyroid.
transformedThyroidData3 = thyroidData.drop('query_on_thyroxine', axis=1)
transformedThyroidData2 = transformedThyroidData3.drop('query_hypothyroid', axis=1)
transformedThyroidData = transformedThyroidData2.drop('query_hyperthyroid', axis=1)
transformedThyroidData.head()
```

### Task 4/5:
Convert the general format and data structure of the data source (from
JSON to CSV, from CSV to JSON, from JSON into a SQL database table,
etc. I want the option to convert any source to any target. So, if I get a
CSV as an input, I want the user to choose an output)
The converted (new) file should be written to disk (local file) or written to a
SQL database like SQL Lite


```python
userInput = input('What format would you like this CSV to be converted into?: \nA) JSON \nB) SQL \nC) CSV\n')
```

```python
# preparing output
if (userInput == 'A') or (userInput == 'a'): # writes the transformed data to a JSON file and prints the first ten records. (I set the print limit to ten since the data rate is exceeded due to the dataset's size.)
    transformedThyroidData.to_json(r'/Users/ninij/Desktop/thyroidDF.json', orient = 'records')
    with open('thyroidDF.json', 'r') as data:
        thyroidJSON = json.load(data)
    print(thyroidJSON[:10]) # I am only printing the first 10 records for simplicity, but the new JSON file is written to the disk with all entries. 

elif (userInput == 'B') or (userInput == 'b'): # writes the transformed data to a SQL database using sqlite
    transformedThyroidData.to_sql('thyroidDF', conn, if_exists='replace', index = True)
    cur = conn.cursor()
    for row in cur.execute('SELECT * FROM thyroidDF LIMIT 10'): # limit set to ten for simplicity but the file created has all entries. 
        print(row) 
    
elif (userInput == 'C') or (userInput == 'c'): # writes the transformed data into a csv file
    transformedThyroidData.to_csv(r'/Users/ninij/Desktop/transformedThyroidDF.csv')
    thyroidCSV = pd.read_csv('transformedThyroidDF.csv')
    print(thyroidCSV)

else: # to throw an error if the user input is not a valid selection
    raise Exception ("This is not a valid input.")
```
